## trees on /scavenger

```python
tree_base = 'https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees'
tree_leaf = '/9308ccc8a4c34c5e3a991ee815222a9691c32476'
trees = tree_base + tree_leaf
trees_json = get_json(trees)
trees_json
```

## df of folders

```python
dftree = pd.DataFrame(trees_json)
dffolders = json_normalize(dftree.tree)
dffolders.head()
```

```python
tree_base = 'https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/'
inner_json = []
for index, row in dffolders.iterrows():
    if row['type'] == 'tree':
        trees = tree_base + row['sha']
        inner_json.append(get_json(trees))

inner_json
```

## df of files

```python
dffiles1 = json_normalize(inner_json)

dffiles1.tree
dffiles1.head()
```

## creating tuples of name / sha for scavenger files

```python
scavenger = []
for index in range(3):
    for f in dffiles1.tree:
        if 'scavengerhunt' in f[index]['path']:
            scavenger.append((f[index]['path'],f[index]['sha']))       
```

```python

scavenger = sorted(scavenger, key=lambda tup: tup[0],reverse=False)
print(scavenger)
```

## getting json of scavenger files

```python
tree_base = 'https://api.github.com/repos/ironhack-datalabs/scavenger/git/blobs/'
content_json = []
for (name, sha) in scavenger:
    trees = tree_base + sha
    content_json.append(get_json(trees))
```

## reordering as scavenger list and decoding

```python
content = []
for i,el in enumerate(content_json):
    content.append((base64.b64decode(el['content']),el['sha']))

result = []
for item in scavenger:
    for cont in content:
        if item[1]==cont[1]:
            result.append(cont[0])
            break

joke = ''.join(map(bytes.decode, result))
joke = ' '.join(joke.split())
print(joke)
```
